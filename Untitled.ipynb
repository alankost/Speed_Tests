```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool
import os

# Generate sample data
n = 500
x = np.random.random(size=n) * 100
y = np.random.random(size=n) * 100
colors = np.random.choice(['red', 'green', 'blue', 'yellow', 'purple'], n)
sizes = np.random.randint(10, 30, n)

# Specify output to HTML file with inline resources for better compatibility
output_file("bokeh_plot.html", mode='inline')

# Create a new plot with tools
p = figure(
    title="Interactive Bokeh Scatter Plot",
    width=800,
    height=600,
    tools="pan,wheel_zoom,box_zoom,reset,save",
    toolbar_location="above"
)

# Add scatter plot (using scatter instead of deprecated circle)
p.scatter(
    x, y,
    size=sizes,
    color=colors,
    alpha=0.6,
    line_color="navy",
    line_width=1,
    legend_label='Data Points'
)

# Add hover tool
hover = HoverTool(tooltips=[
    ("Index", "$index"),
    ("(X,Y)", "($x, $y)")
])
p.add_tools(hover)

# Customize plot
p.title.text_font_size = "16pt"
p.title.align = "center"
p.xaxis.axis_label = "X Axis"
p.yaxis.axis_label = "Y Axis"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"

# Add a line plot on the same figure
x_line = np.linspace(0, 100, 100)
y_line = 50 + 20 * np.sin(x_line / 5)
p.line(x_line, y_line, line_width=3, color='orange', alpha=0.8, legend_label='Sine Wave')

# Configure legend
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Show the plot (this will open in browser)
show(p)

# Print confirmation
print(f"Plot has been generated: {os.path.abspath('bokeh_plot.html')}")
print("The plot should open in your browser automatically.")
print("If not, drag the HTML file into your browser window.")
```
