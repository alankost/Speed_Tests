```python
import cv2
import matplotlib.pyplot as plt

def test_spyder_display():
    print("=== Spyder-Specific Display Test ===")
    
    # Check matplotlib backend
    import matplotlib
    print(f"Matplotlib backend: {matplotlib.get_backend()}")
    
    # Try different display methods
    cap = cv2.VideoCapture(0, cv2.CAP_ANY)
    if not cap.isOpened():
        print("Cannot open camera")
        return
    
    ret, frame = cap.read()
    cap.release()
    
    if not ret or frame is None:
        print("Cannot capture frame")
        return
    
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Method 1: Standard show
    print("Method 1: Standard plt.show()")
    plt.figure(figsize=(6, 4))
    plt.imshow(frame_rgb)
    plt.title('Method 1')
    plt.show()
    
    # Method 2: Force new figure
    print("Method 2: New figure")
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_subplot(111)
    ax.imshow(frame_rgb)
    ax.set_title('Method 2')
    plt.show()
    
    # Method 3: Direct display in console
    print("Method 3: Try to display inline")
    %matplotlib inline  # This only works in IPython/Jupyter
    plt.figure(figsize=(6, 4))
    plt.imshow(frame_rgb)
    plt.title('Method 3 - Inline')
    plt.show()

# Run Spyder test
test_spyder_display()
```
