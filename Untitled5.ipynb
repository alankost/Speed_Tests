```python
import sys
import cv2
import numpy as np
from scipy.optimize import curve_fit

def gaussian_2d(coords, amplitude, x0, y0, sig_x, sig_y, offset):
    """2D Gaussian function."""
    xc, yc = coords
    return offset + amplitude * np.exp(
        -((xc - x0)**2 / (2 * sig_x**2) + (yc - y0)**2 / (2 * sig_y**2))
    )

def fit_gaussian(region):
    """Fit a 2D Gaussian to a region of interest. Returns (sigma_x, sigma_y, success)."""
    h, w = region.shape
    xv = np.arange(w)
    yv = np.arange(h)
    xv, yv = np.meshgrid(xv, yv)
    
    # Initial guesses
    offset_guess = float(np.min(region))
    amplitude_guess = float(np.max(region)) - offset_guess
    x0_guess, y0_guess = w / 2, h / 2
    sigma_x_guess, sigma_y_guess = w / 4, h / 4
    
    try:
        popt, _ = curve_fit(
            gaussian_2d, 
            (xv.ravel(), yv.ravel()), 
            region.ravel(),
            p0=[amplitude_guess, x0_guess, y0_guess, sigma_x_guess, sigma_y_guess, offset_guess],
            bounds=([0, 0, 0, 1, 1, 0], [65535, w, h, w, h, 65535]),
            maxfev=1000
        )
        return abs(popt[3]), abs(popt[4]), True
    except (RuntimeError, ValueError):
        return 0, 0, False

# Use DirectShow backend for faster initialization on Windows
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)

# Reduce buffer size to speed things up
cap.set(cv2.CAP_PROP_BUFFERSIZE, 1)

if not cap.isOpened():
    print("Error: Could not open webcam")
    sys.exit(1)

print("Press 'q' to quit")

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Error: Can't receive frame. Exiting...")
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (11, 11), 0)
    _, thresh = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        circle_result = cv2.minEnclosingCircle(largest_contour)
        center = circle_result[0]
        cx = center[0]
        cy = center[1]
        radius = circle_result[1]
        
        # Extract ROI around the bright spot for Gaussian fitting
        roi_size = int(radius * 3)
        cx_int, cy_int = int(cx), int(cy)
        x1 = max(0, cx_int - roi_size)
        y1 = max(0, cy_int - roi_size)
        x2 = min(gray.shape[1], cx_int + roi_size)
        y2 = min(gray.shape[0], cy_int + roi_size)
        
        roi = gray[y1:y2, x1:x2].astype(float)
        
        if roi.size > 0 and roi.shape[0] > 5 and roi.shape[1] > 5:
            sigma_x, sigma_y, success = fit_gaussian(roi)
            
            if success:
                # 1/e² width = 2 * sigma * 2 = 4 * sigma (full width)
                # 1/e² radius = 2 * sigma
                w_x = 4 * sigma_x  # Full 1/e² width in x
                w_y = 4 * sigma_y  # Full 1/e² width in y
                w_avg = (w_x + w_y) / 2  # Average 1/e² width
                
                cv2.putText(frame, f"1/e^2 width X: {w_x:.1f} px", (10, 150), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
                cv2.putText(frame, f"1/e^2 width Y: {w_y:.1f} px", (10, 180), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
                cv2.putText(frame, f"1/e^2 width avg: {w_avg:.1f} px", (10, 210), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
                
                # Draw 1/e² ellipse
                cv2.ellipse(frame, (cx_int, cy_int), (int(w_x/2), int(w_y/2)), 
                           0, 0, 360, (255, 255, 0), 2)
        
        cv2.circle(frame, (int(cx), int(cy)), int(radius), (0, 255, 0), 2)
        cv2.circle(frame, (int(cx), int(cy)), 5, (0, 0, 255), -1)
        
        cv2.putText(frame, f"Center: ({int(cx)}, {int(cy)})", (10, 30), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        cv2.putText(frame, f"Contour radius: {radius:.1f} px", (10, 60), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    
    cv2.imshow('Webcam Feed', frame)
    
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
